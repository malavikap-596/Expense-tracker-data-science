```python
plt.figure(figsize=(6,6))

cat_data.plot(kind='pie', autopct='%1.1f%%')
plt.title("Category-wise Expense Distribution")

plt.savefig("../images/category_pie.png")
plt.show()
```

```python
plt.figure(figsize=(10,5))

month_data.plot(kind='bar')
plt.title("Monthly Expense Trend")
plt.xticks(rotation=45)

plt.savefig("../images/monthly_bar.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df['amount'], bins=30, kde=True)
plt.title("Spending Distribution")

plt.savefig("../images/spending_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x='category', y='amount', data=df)
plt.xticks(rotation=45)
plt.title("Category vs Expense")

plt.savefig("../images/category_boxplot.png")
plt.show()
```

```python
top_category = cat_data.idxmax()
top_value = cat_data.max()

print("💡 INSIGHTS:\n")

print(f"👉 Highest spending category: {top_category} (₹{top_value})")

# Monthly trend insight
trend = month_data.diff()

if trend.iloc[-1] > 0:
    print("👉 Expenses are increasing 📈")
else:
    print("👉 Expenses are decreasing 📉")

# High spending alert
if top_value > 0.3 * df['amount'].sum():
    print(f"⚠️ High spending detected in {top_category}")
```
